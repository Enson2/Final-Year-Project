**2. Function that uses `hyperopt` library to perform `Bayesian Model-Based Optimisation` for hyperparameter optimization, with default:**
- cv = `Stratified5Fold` (5-fold)
- score = `'recall'`
- default Epochs = `30`


- Use `set_params()` to manually set optimal / other hyperparameters. For example, clf = clf.set_params(**params / random_state=88)

```python
# from hyperopt import hp, tpe, fmin, Trials, space_eval
# from sklearn.model_selection import cross_val_score, StratifiedKFold 
# import numpy as np
# import time

def BayesianOpt_TPE(classifier, space, X_train, y_train, score='recall', num_explore=30):
    
    # Start tracking time
    start_time = time.time()  
    
    # set seed for reproducibility
    rstate = np.random.default_rng(42)
    
    # Define the objective function
    def objective(params):
        clf = classifier.set_params(**params)
        stratified_cv = StratifiedKFold(n_splits = 5)
        cv_score = cross_val_score(clf, X_train, y_train, cv=stratified_cv, scoring=score)
        avg_score = np.mean(cv_score)
        # we wants to maximize score = minimize (-) score
        return -avg_score

    # Perform the search
    trials = Trials()
    best = fmin(fn=objective, space=space, algo=tpe.suggest,
                max_evals=num_explore, trials=trials, rstate=rstate)

    # Calculate time taken
    end_time = time.time()
    time_taken = end_time - start_time
    
    # Print results
    best_params = space_eval(space, best)
    best_loss = -trials.best_trial['result']['loss']
    print("Time taken:", time_taken, "seconds")
    print("Best tuned hyperparameters:")
    print("\t%s" % best_params)
    print("Best", score, "=", best_loss)
    
    # Return the model with its optimal hyperparameters
    return classifier.set_params(**best_params)
```

- This function uses the **`Tree-structured Parzen Estimator (TPE)`** algorithm to perform hyperparameter optimization on the classifier based on the given search space and training data. The function returns the classifier with its optimal hyperparameters.

- User should at first provide wider range of hyperparameters for the algorithm to explore the space and find a good region **(exploration)**. Then, narrow down the range and repeat the process, so that the algorithm can exploit that region and find the optimal hyperparameters **(exploitation)**.

**3. Function to evaluate the model by displaying `classification report`, `confusion matrix`, `Precision-Recall curve`, `ROC curve`**

```python
# import numpy as np
# import matplotlib.pyplot as plt

# from sklearn import metrics
# from sklearn.metrics import balanced_accuracy_score, fbeta_score, make_scorer, matthews_corrcoef
# from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
# # from sklearn.metrics import roc_curve, roc_auc_score
# from sklearn.metrics import PrecisionRecallDisplay

# from IPython.display import display

def model_performance(classifier, y_test, y_predict, y_predict_proba, f_beta=1):
    
    classifier_name = classifier.__class__.__name__
    classes = np.array(["Non-Default", "Default"])
    
    print(f"{classifier_name}'s classification report:\n") # :\n is used to separate outputs
    print(classification_report(y_test, y_predict, target_names=classes))
    
    cm = confusion_matrix(y_test, y_predict)
    display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
    display.plot()
    plt.title(f"{classifier_name}'s confusion matrix")
    plt.show()
    
    print(f"{classifier_name}'s overall:\n")
    print("1. Balanced accuracy score = {:.2f} %".format(balanced_accuracy_score(y_test, y_predict)*100))
    print("2. F{} score = {:.4f}".format(f_beta, fbeta_score(y_test, y_predict, beta=f_beta)))
    print("3. MCC = {:.4f}".format(matthews_corrcoef(y_test, y_predict)))
    
    # Commands to plot Precision-Recall curve 
    PrecisionRecallDisplay.from_estimator(classifier, X_test, y_test, color="orange")
    baseline = len(y_test[y_test==1]) / len(y_test)
    plt.plot([0, 1], [baseline, baseline], linestyle='--', label='Baseline')
    plt.legend(loc='upper right') # "best" will determine the location itself
    plt.title('Precision-Recall curve for the model')
    plt.show()
    
    # Commands to plot ROC cuve
    classifier_name = classifier.__class__.__name__
    y_predict_proba = y_predict_proba[:,1]
    # calculate false and true positive rate
    fpr, tpr, _ = metrics.roc_curve(y_test,  y_predict_proba)
    auc = metrics.roc_auc_score(y_test, y_predict_proba)

    plt.plot(fpr,tpr, color = "darkblue", label = classifier_name + " (area = %0.2f)" % auc)
    plt.plot([0, 1], [0, 1], 'k--', label = "Baseline curve (area = 0.50)")
    plt.xlim([0.0, 1.0]); plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
    plt.title('ROC curve for the model'); plt.legend(loc="lower right")
    plt.show()
```

# **Further Exploration (Class Weights)**

In this section, I will use Decision Tree to investigate how the model performance will be affected given different class weights. This is because once we have found the acceptable class weights using Decision Tree, we can then apply it to the ensemble models like Random Forest. 

I will try manipulating the class weight from 1 to 8 (step length of 0.5) for Default class and fix the class weight for Non-Default class. In order to be fair when comparing the model performances, I will optimize the hyperparameters for each model using the same range of hyperparameters.   

```python
from sklearn.metrics import precision_recall_fscore_support
```

## Manipulate class weights

```python
# creating empty list to store precision, recall, f2 for each class
precision_c0 = []; recall_c0 = []; f2score_c0 = []
precision_c1 = []; recall_c1 = []; f2score_c1 = []

default_weight = np.arange(1,8.5, 0.5)
```

```python
for w in default_weight:
    
    costdt_param = {
        'max_depth': hp.choice('max_depth', np.arange(2, 26)),
        'min_samples_split': hp.quniform('min_samples_split', 0.0001, 0.001, 0.0001)
    }
    
    # hyperparameter tuning and training
    costdt = DecisionTreeClassifier(random_state = seed, class_weight = {0:1,1:w})
    costdt_clf = BayesianOpt_TPE(classifier = costdt, space = costdt_param, 
                                 X_train = X_train, y_train = y_train, 
                                 score = 'recall', num_explore = 20)
    costdt_clf.fit(X_train, y_train)
    cost_y_pred = costdt_clf.predict(X_test)
    
    pre_c0, rec_c0, f2_c0, _ = precision_recall_fscore_support(y_test, cost_y_pred, beta=2, labels=[0])
    pre_c1, rec_c1, f2_c1, _ = precision_recall_fscore_support(y_test, cost_y_pred, beta=2, labels=[1])
    
    # append results into the list
    precision_c0.append(pre_c0); recall_c0.append(rec_c0); f2score_c0.append(f2_c0)
    precision_c1.append(pre_c1); recall_c1.append(rec_c1); f2score_c1.append(f2_c1)
```

We will now derive a linear equation that can be used to find the class weight based on the minority class's proportion in the dataset and plot it

```python
class_weight = [1, 4.5]
class_proportion = [50, 22]

slope, intercept = np.polyfit(class_proportion, class_weight, deg=1)
print(slope); print(intercept)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(class_proportion, class_weight, color='k')

x = np.linspace(2,50,50)
y = slope*x + intercept
plt.plot(x, y, color='b', label='W = -0.125*P + 7.25')
plt.xlim([0,58]); plt.ylim([0,8])
plt.xlabel("Minority Class Proportion (%) - P"); plt.ylabel("Class Weight - W")
plt.title("Scatterplot of Minority Class Proportion and Class Weight")
plt.legend()

plt.show()
```
