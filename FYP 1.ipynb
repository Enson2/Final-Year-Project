# Clean the Data

## Remove "Repayment Status" features

```python
# remove repayment status columns since we are not clear about their definitions
clients_data.drop(['PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6'],
                  axis=1, inplace = True)
```

## Replace unknown values in "Education" with NaN

```python
# Education's feature values after replacing 0,4,5,6 with NaN
education_unknowns = ((clients_data.EDUCATION == 0) | (clients_data.EDUCATION == 5) | 
                      (clients_data.EDUCATION == 6) | (clients_data.EDUCATION == 4))
clients_data.loc[education_unknowns, 'EDUCATION'] = np.NaN

clients_data.EDUCATION.value_counts()
```

## Replace unknown values in "Marital Status" with NaN

```python
# Marital Status's feature values after replacing 0, 3 with NaN
marriage_unknowns = (clients_data.MARRIAGE == 0) | (clients_data.MARRIAGE == 3)
clients_data.loc[marriage_unknowns, 'MARRIAGE'] = np.NaN

clients_data.MARRIAGE.value_counts()
```

## Log-transformation on money values

```python
# function to do log transformation
def findlog(x):
    if x > 0:
        log = np.log(x)
    elif x < 0:
        log = np.log(x*-1)*-1
    elif x == 0:
        log = 0
    return log

col_to_Log = ["LIMIT_BAL","BILL_AMT1","BILL_AMT2","BILL_AMT3",
              "BILL_AMT4","BILL_AMT5","BILL_AMT6",
              "PAY_AMT1","PAY_AMT2","PAY_AMT3","PAY_AMT4","PAY_AMT5","PAY_AMT6"]

for col in col_to_Log:
    clients_data.loc[:,col] = clients_data.loc[:,col].apply(findlog)
```

## Features

### Credit Limit (Feature 1)

```python
# Conditional distribution of Credit Limit given Status of Default

plt.figure(figsize=[6,6])
sns.kdeplot(data= clients_ND, x = 'LIMIT_BAL')
sns.kdeplot(data= clients_D, x = 'LIMIT_BAL')
plt.legend(["Non-Default" , "Default"], loc = "upper right")
plt.ticklabel_format(style='plain')
plt.show()
```

```python
# M-W U test & KS two-samples test (Credit Limit)

print(stats.ks_2samp(clients_ND["LIMIT_BAL"],clients_D["LIMIT_BAL"]))

print(stats.mannwhitneyu(clients_ND["LIMIT_BAL"],clients_D["LIMIT_BAL"]))
```

### Gender (Feature 2)

```python
# Bar graph of Gender given Status of Default

plt.figure(figsize = (10,5))
plt.title('Gender (Male = 1, Female = 2)')
sns.countplot(x = "SEX", data = clients_data, hue = "Default.Payment_Next.Month")
plt.show()
```

```python
# Contigency table (Gender)

gender_contigency_table = pd.crosstab(clients_data["Default.Payment_Next.Month"], 
                                      clients_data["SEX"])
gender_contigency_table.columns = ["Male","Female"]
gender_contigency_table
```

```python
# Chi-Square test of independence (Gender)

(t_stats, p_val, d_f, exp_tab) = stats.chi2_contingency(gender_contigency_table)
print(f"test_statistics = {t_stats}")
print(f"p-values = {p_val}")
```

# Further Analysis

From the Overall Scatter Plot for `Bill` vs `Payment`, we observed that there are a lot of default clients in the bottom area which make sense as we know that when clients with high bill amounts in the current month makes lower payments in the following month, they are more likely to default on payment in next month.

However, we also observed that there are quite number of clients being classified as non-default clients in the bottom area, which might cause some errors to our prediction model.

Therefore, for clients that have high bill amounts and low payment amounts, let us investigate their characteristics to see if there are variables that could help us in distinguishing non-default clients and default clients. 

```python
# extract rows that satisfy these conditions
lowpay = ((clients_data.PAY_AMT1 > 5) & (clients_data.PAY_AMT2 > 5) & 
          (clients_data.PAY_AMT3 > 5) & (clients_data.PAY_AMT4 > 5) & 
          (clients_data.PAY_AMT5 > 5) & (clients_data.PAY_AMT6 > 5))
highbill = ((clients_data.BILL_AMT1 > 5) & (clients_data.BILL_AMT2 > 5) & 
            (clients_data.BILL_AMT3 > 5) & (clients_data.BILL_AMT4 > 5) & 
            (clients_data.BILL_AMT5 > 5) & (clients_data.BILL_AMT6 > 5))
red_region_clients = clients_data[lowpay & highbill]
red_region_clients
```

```python
red_region_clients.describe()
```

```python
# create dataset for non-default clients and default clients given low payment

# red region non-default clients' dataframe
red_region_clients_ND = (
    red_region_clients[red_region_clients["Default.Payment_Next.Month"] == 0]) 
# red region default clients' dataframe
red_region_clients_D = (
    red_region_clients[red_region_clients["Default.Payment_Next.Month"] == 1])
```

# Decision Tree (testing)

```python
from sklearn import tree
```

```python
# remove repayment status columns since we are not clear about their definitions
clients_ori_data.drop(['PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6'], 
                      axis=1, inplace = True)
```

```python
# log transformation on the money values

col_to_Log = ["LIMIT_BAL","BILL_AMT1","BILL_AMT2","BILL_AMT3",
              "BILL_AMT4","BILL_AMT5","BILL_AMT6",
              "PAY_AMT1","PAY_AMT2","PAY_AMT3","PAY_AMT4","PAY_AMT5","PAY_AMT6"]

for col in col_to_Log:
    clients_ori_data.loc[:,col] = clients_ori_data.loc[:,col].apply(findlog)
```

```python
# features
features = ["LIMIT_BAL","SEX","EDUCATION","MARRIAGE","AGE",
            "BILL_AMT1","BILL_AMT2","BILL_AMT3","BILL_AMT4","BILL_AMT5","BILL_AMT6",
           "PAY_AMT1","PAY_AMT2","PAY_AMT3","PAY_AMT4","PAY_AMT5","PAY_AMT6"]
X = clients_ori_data[features]
X
```

```python
# response variable
y = clients_ori_data["Default.Payment_Next.Month"]
y
```

```python
# Split data into training set and testing set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=88)
```

```python
# Class label

clients_class = np.array(["0","1"])
clients_class
```

```python
# call out the decision tree classifier class
depth=3
dt = tree.DecisionTreeClassifier(max_depth=depth)
# Train on data
dt.fit(X_train, y_train)    
plt.figure(figsize=(80,30))
_ = tree.plot_tree(dt, feature_names=features, class_names=clients_class, 
                   filled=True, rounded=True)
plt.show()
```

```python
# compare the predicted label with the actual label to get accuracy score

from sklearn.metrics import accuracy_score

y_pred = dt.predict(X_test) 
accuracy_score(y_test, y_pred)
```
